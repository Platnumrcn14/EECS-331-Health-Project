```python
import pandas as pd
```

Below is the code for cleaning the Air quality data from the WHO. First we filter the data down to Location, Dim1, and the values. Next we pivot the data, dividing the Dim1 column into it's 5 different classes taking each row's value with it. Finally we rename each column appropriately.

```python
df = pd.read_excel("air-quality.xlsx", index_col=False)
filtered_df = df[['Country', 'Location', 'FactValueNumeric']]
pivot_df = filtered_df.pivot_table(index='Country', columns='Location', values='FactValueNumeric')
pivot_df.reset_index(inplace=True)
df = pivot_df.rename(columns={'Cities': 'Air Quality Cities'})
df = df.rename(columns={'Rural': 'Air Quality Rural'})
df = df.rename(columns={'Towns': 'Air Quality Towns'})
df = df.rename(columns={'Total': 'Air Quality Total'})
df = df.rename(columns={'Urban': 'Air Quality Urban'})
df1 = df
df1.head()
```

Below is caloric intake differential data from Oxford University. First we filter the data by the year 2019 as the specified year of interest, next we can drop the columns that aren't needed. Finally we rename column 'Entity' to 'Country' to be uniform with other datasets.

```python
df = pd.read_excel("CV-of-caloric-intake-2019.xlsx", index_col=False)
filtered_df = df[df['Year'] == 2019]
df=filtered_df.drop(columns=['Year'])
df = df.rename(columns={'Entity': 'Country'})
df2 = df
df2.head()
```

Below is the cleaning of Child mortality data from the WHO. First step is to filter by the year. Next is to filter down to the Location, Sex, and the child mortality values we want. After that we pivot the data using the country as the index, and sex as the column headers. Finally we rename the columns to with appropriate labels.

```python
df = pd.read_excel("childhood-mortality-under-5-2019.xlsx", index_col=False)
filtered_df = df[df['Year'] == 2019]
filtered_df = filtered_df[['Country', 'Sex', 'FactValueNumeric']]
pivot_df = filtered_df.pivot_table(index='Country', columns='Sex', values='FactValueNumeric')
pivot_df.reset_index(inplace=True)
df3 = pivot_df
df3.head()
```

The data represented in this dataset is the alcohol consumption per capita in 2019 from the WHO. First we filter the data down to the columns of interest. Next pivot the data by sex with the value of each sex for each country. Finally we rename the columns to have appropriate titles.

```python
df = pd.read_excel("alcohol-data.xlsx", index_col=False)
filtered_df = df[['Country', 'Sex', 'FactValueNumeric']]
pivot_df = df.pivot_table(index='Country', columns='Sex', values='FactValueNumeric')
pivot_df.reset_index(inplace=True)
df = pivot_df.rename(columns={'Both sexes': 'Alcohol Both sexes'})
df = df.rename(columns={'Male': 'Alcohol Male'})
df = df.rename(columns={'Female': 'Alcohol Female'})
df4 = df
df4.head()
```

Below is the cleaning for the disease and cause of deaths data from Oxford University. First we filter the data down to only the year 2019. Next drop the unneeded 'Year' column. Finally rename the 'Entity' column to 'Country' to match other datasets.

```python
df = pd.read_excel("disease-2019.xlsx", index_col=False)
df.drop(columns=['Year'], inplace=True)
df = df.rename(columns={'Entity': 'Country'})
df5 = df
df5.head()
```

The life expectancy data from the WHO is already in the format needed it will be uploaded as df6

```python
df6 = pd.read_excel("Life-Expectancy-2019.xlsx", index_col=False)
df6.head()
```

Below is the data containing nutrition information and population metrics from Oxford University. In this data cleaner we filter down the dataset to the year 2019 exclusively and remove all unneeded columns until we are left with the 5 columns of interest below. The data can then be saved to be merged with the other data.

```python
df = pd.read_excel("global-food-protien-fat-2019.xlsx", index_col=False)
filtered_df = df[df['Year'] == 2019]
filtered_df = filtered_df[['Country', 'Population', 'Food supply (kcal per capita per day)', 'Food supply (Protein g per capita per day)','Food supply (Fat g per capita per day)']]
df7 = filtered_df
df7.head()
```

Below is water quality data from Oxford University. First we filter the dataset by the year 2019 exclusively and drop the unneeded 'Year' column. Then we rename the column 'Entity' to 'Country' for merging purposes.

```python
df = pd.read_excel("number-without-improved-water-2019.xlsx", index_col=False)
filtered_df = df[df['Year']== 2019]
df=df.drop(columns=['Year'])
df = df.rename(columns={'Entity': 'Country'})
df8 = df
df8.head()
```

Finally we have Obesity data from the World Obesity Federation website. All we have to do to suit the data is drop the 'Rank' column

```python
df9 = pd.read_excel("Obesity.xlsx", index_col=False)
df9=df9.drop(columns=['Rank'])
df9.head()
```

#### Merging data
Now we merge all of the datasets into one master dataset

```python
df = pd.merge(df1, df2, on='Country')
```

```python
df = pd.merge(df, df3, on='Country')
df = pd.merge(df, df4, on='Country')
df = pd.merge(df, df5, on='Country')
df = pd.merge(df, df6, on='Country')
df = pd.merge(df, df7, on='Country')
df = pd.merge(df, df8, on='Country')
df = pd.merge(df, df9, on='Country')
```

Next those three forgien values are standardized for the next step, weighting the variable to find the one with the most influence

```python
df4 = df2.copy()
scaler = StandardScaler()
columns_to_standardize = ['Air Quality Cities', 'Air Quality Rural', 'Air Quality Total', 'Air Quality Towns', 'Air Quality Urban', 'Alcohol Both sexes', 'Both sexes', 'Coefficient of Variation (CV) of caloric intake', 'Self-harm', 'Exposure to forces of nature', 'Conflict and terrorism', 'Interpersonal violence', 'Neglected tropical diseases and malaria', 'Substance use disorders', 'Skin and subcutaneous diseases', 'Enteric infections', 'Diabetes and kidney diseases', 'Cardiovascular diseases', 'Digestive diseases', 'Nutritional deficiencies', 'Respiratory infections and tuberculosis', 'Neonatal disorders', 'Chronic respiratory diseases', 'Other non-communicable diseases', 'Maternal disorders', 'Unintentional injuries', 'Musculoskeletal disorders', 'Neoplasms', 'Mental disorders', 'Neurological disorders', 'HIV/AIDS and sexually transmitted infections', 'Transport injuries', 'Sense organ diseases', 'Population', 'Food supply (kcal per capita per day)', 'Food supply (Protein g per capita per day)', 'Food supply (Fat g per capita per day)', 'The totalpopulation not using improved water sources', '% Obesity']
df4[columns_to_standardize] = scaler.fit_transform(df4[columns_to_standardize])
df4.head()
```

Next smalls for loop is performed on select variables to find the most heavily weighted of select variables.

```python
holder = df4.iloc[0]['The totalpopulation not using improved water sources']
water_values = []
for i in range(0, 11):
    i = i / 10
    df4.at[0, 'The totalpopulation not using improved water sources'] = i
    water_values.append(health_pipe.predict(df4)[0])
df4.at[0, 'The totalpopulation not using improved water sources'] = holder
d_water = water_values[0] - water_values[1]

holder = df4.iloc[0]['Air Quality Urban']
air_q_values = []
for i in range(0, 11):
    i = i / 10
    df4.at[0, 'Air Quality Urban'] = i
    air_q_values.append(health_pipe.predict(df4)[0])
df4.at[0, 'Air Quality Urban'] = holder
d_air = air_q_values[0] - air_q_values[1]

holder = df4.iloc[0]['Food supply (kcal per capita per day)']
kcal_values = []
for i in range(0, 11):
    i = i/10
    df4.at[0, 'Food supply (kcal per capita per day)'] = i
    kcal_values.append(health_pipe.predict(df4)[0])
df4.at[0, 'Food supply (kcal per capita per day)'] = holder
d_kcal = kcal_values[0] - kcal_values[1]

holder = df4.iloc[0]['Food supply (Protein g per capita per day)']
protein_values = []
for i in range(0, 11):
    i = i / 10
    df4.at[0, 'Food supply (Protein g per capita per day)'] = i
    protein_values.append(health_pipe.predict(df4)[0])
df4.at[0, 'Food supply (Protein g per capita per day)'] = holder
d_protein = protein_values[0] - protein_values[1]

holder = df4.iloc[0]['Food supply (Fat g per capita per day)']
fat_values = []
for i in range(0, 11):
    i = i / 10
    df2.at[0, 'Food supply (Fat g per capita per day)'] = i
    fat_values.append(health_pipe.predict(df4)[0])
df4.at[0, 'Food supply (Fat g per capita per day)'] = holder
d_fat = fat_values[0] - fat_values[1]
```

```python
df3 = pd.DataFrame(columns=['Attribute', 'Derivatives'])
df3.loc[0] = ["Water Quality", d_water]
df3.loc[1] = ["Air Quality", d_air]
df3.loc[2] = ["Calorie Consumption", d_kcal]
df3.loc[3] = ["Protein Consumption", d_protein]
df3.loc[4] = ["Fat Consumption", d_fat]

df3.head()
```

```python
np.random.seed(1234)  # DO NOT CHANGE!

boot6 = pd.DataFrame()
for i in range(0, 6):
    bootstrap_sample = one_sample.sample(n=len(one_sample), replace=True)
    bootstrap_sample['replicate'] = i
    boot6 = pd.concat([boot6, bootstrap_sample], ignore_index=True)
boot6
```

```python
boot6_dist = alt.Chart(boot6, title="Bootstrap Sample Distributions").mark_bar().encode(
    x=alt.X('Tobacco products used per week')
        .title('Tobacco products used per week')
        .bin(maxbins=30),
    y=alt.Y('count()', title='Count of Records')
).properties(
    height=100
).facet(
    row='replicate',
    columns=1
)

boot6_dist
```

```python
boot6_means = boot6.groupby('replicate')['Tobacco products used per week'].mean().reset_index()
boot6_means = boot6_means.rename(columns={'sample': 'mean_tobacco_use'})
boot6_means
```
